## Leitura dos datasets já separados e criação do modelo com melhores parâmetros utilizando todo o dataset de treino.

## CSTR

```python
X_train = joblib.load("../dados/atv3/X_train_CSTR.pkl")
y_train = joblib.load("../dados/atv3/y_train_CSTR.pkl")
X_test = joblib.load("../dados/atv3/X_test_CSTR.pkl")
y_test = joblib.load("../dados/atv3/y_test_CSTR.pkl")
```

```python
results_cstr = pd.read_csv("../dados/atv3/results_gs_CSTR.csv")
results_cstr.head(10)
```

```python
best = results_cstr['mean_test_score'].max()
results_cstr.query(f'mean_test_score == {best}')
```

```python
model_cstr = MultinomialNB(alpha=0.5, fit_prior=False)
```

```python
model_cstr.fit(X_train, y_train)
```

```python
predictions_cstr = model_cstr.predict(X_test)
```

```python
print("F1-Micro: ", f1_score(y_test, predictions_cstr, average='micro', zero_division=0.0))
print("F1-Macro: ", f1_score(y_test, predictions_cstr, average='macro', zero_division=0.0))
print("Accuracy: ", accuracy_score(y_test, predictions_cstr))
```
